# Limits of Diversification

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import pandas as pd
import numpy as np
import edhec_risk_kit as erk
```

```python
ind_return = erk.get_ind_returns()
ind_size = erk.get_ind_size()
ind_nfirms = erk.get_ind_nfirms()
```

```python
ind_size.head()
```

```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap = ind_mktcap.sum(axis=1)
```

```python
total_mktcap.plot(figsize=(12, 6))
```

```python
# divide industry market cap by total market cap for year rowwise
ind_capweight = ind_mktcap.divide(total_mktcap, axis=0)
```

```python
ind_capweight.head()
```

```python
# sum for 1926 to confirm each capweight adds to 1 for the year (per month)
ind_capweight['1926'].sum(axis=1)
```

```python
ind_capweight[['Fin', 'Steel']].plot(figsize=(12, 6))
```

```python
total_market_return = (ind_capweight * ind_return).sum(axis=1)
```

```python
total_market_return.plot()
```

```python
total_market_index = erk.drawdown(total_market_return).wealth
total_market_index.plot(figsize=(10, 5), title='Total Cap Weighted Market Index')
```

```python
# moving average 36 month window of the index
total_market_index['1980':].plot(figsize=(12, 6))
total_market_index['1980':].rolling(window=36).mean().plot(figsize=(12,6))
```

```python
tmi_tr36rets = total_market_return.rolling(window=36).aggregate(erk.annualize_rets, periods_per_year=12)
tmi_tr36rets.plot(figsize=(12, 6), label='Trailing 36 months returns', legend=True)
total_market_return.plot(label='Returns', legend=True)
```

## Rolling Correlation - along with MultiIndexes and `.groupby`

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.index.names = ['date', 'industry']
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())
```

```python
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label='Trailing 36 month returns', figsize=(12, 6), legend=True)
ind_tr36corr.plot(label='Trailing 36 month correlations', figsize=(12, 6), legend=True, secondary_y=True)
```

```python
# only from 2007 onwards
tmi_tr36rets['2007':].plot(label='Trailing 36 month returns', figsize=(12, 6), legend=True)
ind_tr36corr['2007':].plot(label='Trailing 36 month correlations', figsize=(12, 6), legend=True, secondary_y=True)
```

```python
# view correlation of correlation with returns
tmi_tr36rets.corr(ind_tr36corr)
```
